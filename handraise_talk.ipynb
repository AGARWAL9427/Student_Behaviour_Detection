```python
#TO merge
# Install YOLO (if not installed)
#!pip install ultralytics

# Import required libraries
import torch
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import numpy as np



# ✅ Define paths to trained YOLO models in Google Drive
model_paths = [
    "/content/drive/MyDrive/models/TeacherBehaviour60/detect/train/weights/best.pt",
    "/content/drive/MyDrive/models/Teacher60_final/detect/train/weights/best.pt",
    "/content/drive/MyDrive/models/TalkTeacherBehaviour59/detect/train/weights/best.pt",
    "/content/drive/MyDrive/models/TalkModel44/detect/train/weights/best.pt",
    "/content/drive/MyDrive/models/Standing60_final/detect/train2/weights/best.pt",
    "/content/drive/MyDrive/models/HandriseModel60_final/detect/train/weights/best.pt",
    "/content/drive/MyDrive/models/Discuss_Model_final/detect/train/weights/best.pt",
    "/content/drive/MyDrive/models/BlackBoardScreenTeacher_Model_final/detect/train/weights/best.pt",
    "/content/drive/MyDrive/models/BlackBoardScreen_Model_final/detect/train/weights/best.pt"
    # Add paths for all 9 models here
]

# ✅ Load all models
models = [YOLO(path) for path in model_paths]

# ✅ Load test image
image_path = "/content/drive/MyDrive/dataset/SCB5-Handrise-Read-write-2024-9-17/images/train/0089_001798.jpg"  # Change this to your test image
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB

# ✅ Run all models on the image & store detections
for model in models:
    results = model.predict(image, conf=0.25)  # Run YOLOv11 inference
    detections = results[0]  # Get first result

    # ✅ Draw bounding boxes from detected objects
    for box in detections.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box
        class_id = int(box.cls[0])  # Class ID
        label = model.names[class_id]  # Class name
        conf = float(box.conf[0])  # Confidence score

        # Assign random color to each class
        color = tuple(np.random.randint(0, 255, 3).tolist())

        # Draw bounding box & label
        cv2.rectangle(image, (x1, y1), (x2, y2), color, 2)
        cv2.putText(image, f"{label}: {conf:.2f}", (x1, y1 - 5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

# ✅ Display final image with all detections
plt.figure(figsize=(10, 6))
plt.imshow(image)
plt.axis("off")
plt.show()
```
